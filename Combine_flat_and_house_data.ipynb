```
import numpy as np
import pandas as pd
```

```
flats = pd.read_csv('/content/flats_clean_csv_j')
house = pd.read_csv('/content/house_clean_csv_j')
```

```
#concat flats and house clean dataset
df = pd.concat([flats,house],ignore_index=True)
```

```
#shuffle all combine data
df = df.sample(df.shape[0],ignore_index=True)
```
